```python
from os import chdir as cd
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt


pathin = '/Users/ethan/Documents/GitHub/LaCoBra/2021/Data/'
file = 'AdamSandler.csv'

cd(pathin)

df = pd.read_csv(file)
df.head()
```

```python
ax = sns.regplot(x = 'Freshness', y = 'BoxOffice', data = df)
for i, val in enumerate(df['MovieTitle']):
    ax.annotate(val, (df.Freshness.iat[i],df.BoxOffice.iat[i]))

ax.set(title = 'Do better Adam Sandler movies make more money?', xlabel = 'RT Freshness', ylabel='Box Office $M')
sns.despine(top=True, right=True)
sns.set_context("poster", font_scale=.5)


plt.savefig('/Users/ethan/Documents/GitHub/LaCoBra/2021/Resources/Sandler.png', facecolor='white', bbox_inches='tight')
```
